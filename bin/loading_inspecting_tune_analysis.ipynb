## Demonstration of how to interact with archived Ray Tune analysis results

Ray Tune results are saved to ~/.ray_results by default or a custom location via: tune.run(local_dir="...", upload_dir="remote_version_using_s3_or_gs_link").  They can be reloaded below as analysis objects if we know their location.  Not sure how well this works if using remote storage

```python
import ray
from ray import tune
```

Each time we call tune.run() it creates a new experiment(?) in the storage directory.  This creates an experiment_state file:

- ./TrainableName/experiment_state-yyyy-mm-dd_hh-mm-ss.json

And a separate subdir for each trial:

- ./TrainableName/TrainableName_trialid_yyyy-mm-dd_hh-mm-ss_SomeOtherId

We can reload the analysis data by loading the experiment_state json file

```python
analysis = tune.ExperimentAnalysis("../mock_data/untracked/ray_results/MyTrainable/experiment_state-2020-02-27_08-04-11.json")
```

The results are available, as we can see from the dataframe() trial_dataframes attributes 

```python
analysis.dataframe()
```

```python
analysis.trial_dataframes
```
